```python
setosa = []
versicolor = []
virginica = []
for i in df["target"]:
  if i == 0:
    setosa.append(1)
    versicolor.append(0)
    virginica.append(0)
  elif i ==1:
    setosa.append(0)
    versicolor.append(1)
    virginica.append(0)
  else:
    setosa.append(0)
    versicolor.append(0)
    virginica.append(1)

df["is setosa"] = setosa
df["is versicolor"] = versicolor
df["is virginica"] = virginica
```
